# Create Segment and Plot

```python
#output_dir = './data_v3/train/'
output_dir = './data_v3/old_test/'
#output_dir = './data_v3/new_test/'
```

```python
from tqdm import tqdm_notebook
from collections import namedtuple
from shapely.geometry import Point, LineString
import time
import gc
import matplotlib
import matplotlib.pyplot as plt
import tracemalloc

XMIN = 120
XMAX = 122
YMIN = 21.5
YMAX = 25.5

def featureExtraction(dataX, dataY, time_steps, step):
    
    f, ax = plt.subplots(1, 1)
    f.set_figheight(5)    # 32 pixel
    f.set_figwidth(5)     # 32 pixel
    
    counter = 1
    
    first = True
    now = time.time()
#     segments = []
#     labels_tt = []
#     labels_splv_hard = []
#     labels_splv_simple = []
    i = 0
    pbar = tqdm_notebook(total=len(dataX)-time_steps)
    while i <= len(dataX)-time_steps: # user不一樣的情況
        #tracemalloc.start()
        
        if counter%50==0:
            gc.collect()
#    for i in range(0, len(dataX)-time_steps, step): # user不一樣的情況
        #print(i)
        stop = False
        users = dataX['user'].values[i: i+time_steps]
        if(first):
            for index in range(len(users)):
                if users[index]!=users[0]:
                    #print(i, i+index, 'user' )
                    i = i+index
                    pbar.update(index)
                    stop = True
                    break
                first = False
                
        else:
#            for index in range(step, len(users)):
            for index in range(len(users)):
                if users[index]!=users[0]:
                    #print(i, i+index, 'user' )
                    i = i+index
                    pbar.update(index)
                    stop = True
                    break
        if (stop):
            continue

        time_ = dataX['time'].values[i:i+time_steps]
        for index in range(len(time_)-1): # 時間超過15min的情況
            #if (time_[index+1] - time_[index]) > (15*60000): # 15min-> ms
            if ((time_[index+1] - time_[index]) > 5000) or ((time_[index+1] - time_[index]) < 1000):
                i = i+index+1
                pbar.update(index+1)
                stop = True
                break

        if (stop):
            continue
        
        
        currCounty = dataX['county'].values[i]
        clat = dataX['cLat'].values[i: i+time_steps]
        clng = dataX['cLng'].values[i: i+time_steps]
        
        #label_tt = dataY['label'][i: i+time_steps]
        label_tt = dataY['label'].values[i]
        
        cell_loc_cnt = {}
        tmp_p_list = zip(clng, clat)
        for _tmp_p in tmp_p_list:
            if _tmp_p not in cell_loc_cnt:
                cell_loc_cnt[_tmp_p] = 0
            cell_loc_cnt[_tmp_p] += 1
               
#         p_list = [Point(k) for k in cell_loc_cnt.keys()]
#         plot_traj = True
#         try:
#             plot_traj = True
#             traj = LineString(p_list)
#         except ValueError as e:
#             plot_traj = False
        
        #print(cell_loc_cnt)
        
#         print(currCounty)
#         print('label: ', label_tt)
        
        '''Plot'''
        
        countyInfo = county.loc[county['COUNTYNAME']==currCounty]        
        maxx = min(XMAX,countyInfo['maxx'].values[0])
        minx = max(XMIN,countyInfo['minx'].values[0])
        maxy = min(YMAX, countyInfo['maxy'].values[0])
        miny = max(YMIN, countyInfo['miny'].values[0])
        
        lenx = maxx - minx
        leny = maxy - miny
        square_len = max(lenx, leny)
        
        ax.set_xlim((minx, minx+square_len))
        ax.set_ylim((miny, miny+square_len))
#         if plot_traj==True:
#             gpd.GeoSeries(traj).plot(ax=ax, linewidth=10, color='gray', alpha=0.1)
        
        # plot cell resident time
        for key in cell_loc_cnt:
            ps = gpd.GeoSeries(Point(key))
            cmap = matplotlib.cm.get_cmap('Purples')
            _color = cmap(0.3+(cell_loc_cnt[key]/time_steps)*0.5)   # use [0.3,0.8] of colormap
            ps.plot(ax=ax, marker='o', color=_color, markersize=20, alpha=0.3)

        # 需要在圖上加點的話可以參考這邊
        p1 = gpd.GeoSeries(countyInfo['centroid'])
        p1.plot(ax=ax, marker='o', color='red', markersize=50)
        # p2 = gpd.GeoSeries(Point(120.99472986085, 24.90748396855956))
        # p2.plot(ax=ax, marker='o', color='green', markersize=20)
        
        plt.axis('off')
        plt.savefig(output_dir+"%s/%s_%s.png" % (label_tt, currCounty, counter), bbox_inches='tight')
        plt.cla()
        
        counter += 1
        pbar.update(step)
        i += step
```

```python
import gc
STEP_DISTANCE = 1
for ss in [80]:
    TIME_STEPS = ss
    featureExtraction(df_x, df_y, TIME_STEPS, STEP_DISTANCE)
    
    gc.collect()
```
